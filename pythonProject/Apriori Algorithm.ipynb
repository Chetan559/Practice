#  Basic Apriori Algorithm

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Load the MovieLens dataset
file_path = r"C:\Users\Chetan\Downloads\ml-latest-small\ml-latest-small\ratings.csv"
df = pd.read_csv(file_path)

# Convert ratings into a binary format (liked/disliked)
df['liked'] = df['rating'].apply(lambda x: x >= 3.5)


# Create a pivot table for transactions (users as rows, movies as columns)
basket = df.pivot_table(index='userId', columns='movieId', values='liked', fill_value=False).astype(bool)

# Apply the Apriori algorithm
frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True, low_memory=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Display the first few association rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

# Improved Apriori (FP-Growth Algorithm)

```python
from mlxtend.frequent_patterns import fpgrowth

# Apply the FP-Growth algorithm (Improved Apriori)
frequent_itemsets_fpgrowth = fpgrowth(basket, min_support=0.05, use_colnames=True)

# Generate association rules for FP-Growth
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="lift", min_threshold=1.0)

# Display the first few association rules
print(rules_fpgrowth[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```
